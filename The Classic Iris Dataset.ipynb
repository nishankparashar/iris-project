# Prepare our data for modeling
This involves splitting the data bac out into plain Numpy arrays

```python
X_train = df_train.drop(columns=["target", "target_name"]).values
Y_train = df_train["target"].values
```

# Modeling - What is our baseline?

What is the simplest data model can we think of?

In this case, if our base line model is just randomly gussing the species of the flower, or guessing a single species for the every data point, we woiuld expect to have accuracy of 0.33 or 33%, since we have 3 different classes that are evenly balanced.

So our model should at least beat 33% accuracy.

# Modeling - Simple Manual Model

Let's manually look at our data and decide some cutoff points for classification.

```python
def single_feature_prediction(petal_length):
    """Predicts the Iris species given the petal lenght."""
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.6:
        return 1
    else:
        return 2
```

```python
X_train[:, 2]
```

```python
manual_y_prediction = [single_feature_prediction(val) for val in X_train[:, 2]]
```

```python
manual_model_accuracy = np.mean(manual_y_prediction == Y_train)
```

```python
print(f"Manual Model Accuracy: {manual_model_accuracy * 100:.2f} %")
```

# Modeling Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

## Using a validationset to evaluate our model

```python
model = LogisticRegression()
```

```python
Xt, Xv, yt, yv = train_test_split(X_train, Y_train, test_size = 0.25)
```

```python
Xt.shape
```

```python
Xv.shape
```

```python
model.fit(Xt, yt)
```

```python
y_pred = model.predict(Xv)
```

```python
model_accuracy = model.score(Xv, yv)
print(f"Model Accuracy: {model_accuracy * 100:.2f} %")
```

## Using crossvalidation to evaluate our model

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
model = LogisticRegression(max_iter=200)
```

```python
accuracies = cross_val_score(model, X_train, Y_train, cv=5, scoring = "accuracy")
```

```python
np.mean(accuracies)
```

## Where are we misclassfiying points

```python
y_pred = cross_val_predict(model, X_train, Y_train, cv=5)
```

```python
predicted_correctly_mask = y_pred == Y_train
```

```python
not_predicted_correctly = ~predicted_correctly_mask
```

```python
X_train[not_predicted_correctly]
```

```python
df_prediction = df_train.copy()
```

```python
df_prediction["correct_prediction"] = predicted_correctly_mask
```

```python
df_prediction["prediction"] = y_pred
```

```python
df_prediction["prediction_label"] = df_prediction["prediction"].map({0: "setosa", 1: "versicolor", 2: "virginica"})
```

```python
df_prediction.head()
```

```python
sns.scatterplot(x="petal length (cm)", y="petal width (cm)", hue="target_name", data=df_prediction)
```

```python
def plot_incorrect_prediction(df_prediction, x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots(2, 2, figsize=(10,10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="prediction_label", data=df_prediction, ax=axs[0])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="target_name", data=df_prediction, ax=axs[1])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="correct_prediction", data=df_prediction, ax=axs[2])
    axs[3].set_visible(False)
    plt.show()
```

```python
plot_incorrect_prediction(df_prediction, "petal length (cm)", "petal width (cm)")
```

# Model Tuning

What is model Tuning?

Model Tuning is tryingto determine the parameters of your model (these are also known as "hyperparameters") that maximize the model performance.

```python
for reg_param in (0.1, 0.3, 0.7, 1.1, 1.4, 1.9, 2.1):
    print(reg_param)
    model = LogisticRegression(max_iter=200, C=reg_param)
    accuracies = cross_val_score(model, X_train, Y_train, cv=5, scoring = "accuracy")
    print(f"Model Accuracy: {np.mean(accuracies) * 100:.2f} %")
```

# Final Model

```python
model = LogisticRegression(max_iter=200, C=1)
```
